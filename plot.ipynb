# Plotting distribution of all scores

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.ticker import MaxNLocator
```

```python
#score = ['0-0', '1-0 | 0-1', '1-1','2-0 | 0-2', '1-0 | 0-1', '1-0 | 0-1', '2-0']
score = []
score.append('2-0 | 0-2') # Catar 2 Equador 0
score.append('6-2 | 2-6') # Inglaterra 6 Iran 2
score.append('2-0 | 0-2') # Senegal 0 Paises Bajos 2
score.append('1-1')       # Estados Unidos 1 Gales 1
score.append('2-1 | 1-2') # Argentina 1 Arabia Saudita 2
score.append('0-0')       # Dinamarca 0 Tunez 0
score.append('0-0')       # Mexico 0 Polonia 0
score.append('4-1 | 1-4') # Francia 4 Australia 1
score.append('0-0')       # Marruecos 0 Croacia 0
score.append('2-1 | 1-2') # Alemania 1 Japon 2
score.append('7-0 | 0-7') # Espana 7 Costa Rica 0
score.append('1-0 | 0-1') # Belgica 1 Canada 0
score.append('1-0 | 0-1') # Suiza 1 Camerun 0
score.append('0-0')       # Uruguay 0 Corea del Sur 0
score.append('3-2 | 2-3') # Portugal 3 Ghana 2
score.append('2-0 | 0-2') # Brasil 2 Serbia 0
score.append('2-0 | 0-2') # Gales 0 Iran 2
score.append('3-1 | 1-3') # Catae 1 Senegal 3
score.append('1-1')       # Paises Bajos 1 Ecuador 1
score.append('0-0')       # Inglaterra 0 Estados Unidos 0
score.append('1-0 | 0-1') # Tunez 0 Australia 1
score.append('2-0 | 0-2') # Polonia 2 Aurabia Saudita 0
score.append('2-1 | 1-2') # Francia 2 Dinamarca 1
score.append('2-0 | 0-2') # Argentina 2 México 0
score.append('1-0 | 0-1') # Japón 0 Costa Rica 1
score.append('2-0 | 0-2') # Bélgica 0 Marruecos 2
score.append('4-1 | 1-4') # Croacia 4 Canadá 1
score.append('1-1')       # Espana 1 Alemania 1
```

```python
fig, ax = plt.subplots(figsize=(13,8))# figsize=(10,8)
plt.title('Histograma de los resultados de la FIFA World Cup Qatar 2022', fontsize=15)

ax.bar(*np.unique(score, return_counts=True),  color='blue',  alpha=0.5, label='Ronda 1 y 2 de la fase de grupos')
ax.tick_params(labelsize=12)
ax.legend(fontsize=12)
ax.set_xlabel('Resultados de partidos de futbol', fontsize=12)
ax.set_ylabel('Numero de partidos', fontsize=12)
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
ax.grid()
#plt.show()
plt.savefig('plot.png')
```
